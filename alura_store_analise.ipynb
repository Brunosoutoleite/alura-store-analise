### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Lendo os arquivos CSV das quatro lojas
loja1 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv")
loja2 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv")
loja3 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv")
loja4 = pd.read_csv("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv")

faturamento_loja1 = loja1["Preço"].sum()
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

print("Faturamento Loja 1:", faturamento_loja1)
print("Faturamento Loja 2:", faturamento_loja2)
print("Faturamento Loja 3:", faturamento_loja3)
print("Faturamento Loja 4:", faturamento_loja4)




```

# 2. Vendas por Categoria


```python
print("Categorias mais vendidas na Loja 1:")
print(loja1['Categoria do Produto'].value_counts())

print("\nCategorias mais vendidas na Loja 2:")
print(loja2['Categoria do Produto'].value_counts())

print("\nCategorias mais vendidas na Loja 3:")
print(loja3['Categoria do Produto'].value_counts())

print("\nCategorias mais vendidas na Loja 4:")
print(loja4['Categoria do Produto'].value_counts())
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print("Média de avaliação Loja 1:", media_avaliacao_loja1)
print("Média de avaliação Loja 2:", media_avaliacao_loja2)
print("Média de avaliação Loja 3:", media_avaliacao_loja3)
print("Média de avaliação Loja 4:", media_avaliacao_loja4)
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_menos_vendidos(loja, nome_loja):
    print(f"Produtos mais vendidos na {nome_loja}:")
    print(loja['Produto'].value_counts().head(5))
    print(f"\nProdutos menos vendidos na {nome_loja}:")
    print(loja['Produto'].value_counts().tail(5))
    print("\n" + "="*40 + "\n")

produtos_mais_menos_vendidos(loja1, "Loja 1")
produtos_mais_menos_vendidos(loja2, "Loja 2")
produtos_mais_menos_vendidos(loja3, "Loja 3")
produtos_mais_menos_vendidos(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
frete_medio_loja1 = loja1['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

print("Frete médio Loja 1:", frete_medio_loja1)
print("Frete médio Loja 2:", frete_medio_loja2)
print("Frete médio Loja 3:", frete_medio_loja3)
print("Frete médio Loja 4:", frete_medio_loja4)
```

# Resumo

```python
import matplotlib.pyplot as plt

# Dados
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
faturamento = [1534509.12, 1488459.06, 1464025.03, 1384497.58]
avaliacoes = [3.976685, 4.037304, 4.048326, 3.995759]
fretes = [34.691805, 33.621941, 33.074464, 31.278997]

# Criar a figura
fig, axs = plt.subplots(1, 3, figsize=(20, 6))

# Gráfico 1 - Pizza (Faturamento)
axs[0].pie(faturamento, labels=lojas, autopct='R$%.2f', colors=plt.cm.Set3.colors)
axs[0].set_title('Faturamento Total por Loja')

# Gráfico 2 - Linhas (Avaliação)
axs[1].plot(lojas, avaliacoes, marker='o', linestyle='-', color='darkgreen')
axs[1].set_title('Avaliação Média por Loja')
axs[1].set_ylim(3.9, 4.1)
axs[1].grid(True)
for i, val in enumerate(avaliacoes):
    axs[1].text(i, val + 0.005, f"{val:.3f}", ha='center')

# Gráfico 3 - Barras Horizontais (Frete)
bars = axs[2].barh(lojas, fretes, color=plt.cm.coolwarm([0.2, 0.4, 0.6, 0.8]))
axs[2].set_title('Custo Médio do Frete por Loja')
axs[2].set_xlabel('Valor em R$')
for bar in bars:
    width = bar.get_width()
    axs[2].text(width + 0.3, bar.get_y() + bar.get_height()/2, f'R${width:.2f}', va='center')

# Ajustar layout
plt.tight_layout()
plt.show()






```
